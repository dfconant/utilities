```python
%load_ext autoreload
%autoreload 2
from ECoGstats import makeD_multi, smooth_formants, smooth_warp_formants, process_kin, normalize_contour_length
import seaborn as sns
import prettyplotlib as ppl
from matplotlib.backends.backend_agg import (FigureCanvasAgg as FigureCanvas)
from scipy.signal import savgol_filter
import pandas as pd
from sklearn.lda import LDA
```

```python
def gestural_score(Ds,feats,labels,fig_filename,window = np.array([-1,1])):

    co = ppl.brewer2mpl.get_map('Set1','Qualitative',9).mpl_colors
    align_window = window
    tt = np.linspace(align_window[0],align_window[1],Ds[0].shape[1])
    sns.set_context('talk')
    sns.set_style('dark')
    fig, axes = plt.subplots(figsize=(12,40))
    axes.axis('off')
    canvas = FigureCanvas(fig)

    for cnt,f in enumerate(feats):
        yscale = np.empty((len(Ds),2))
        ax = fig.add_subplot(len(feats),1,cnt+1)
        for iD,D in enumerate(Ds):
            mu = savgol_filter(np.nanmean(D[cnt,:,:],axis=1),5,3)
            ste = savgol_filter(np.nanstd(D[cnt,:,:],axis=1)/D.shape[2],5,3)
            yscale[iD,:] = np.array([np.nanmin(mu-2*ste), np.nanmax(mu+2*ste)])
            plt.plot(tt,mu,color = co[iD],label=labels[iD])
            plt.fill_between(tt,mu-ste, mu+ste,color = co[iD],alpha=0.5)
            plt.plot([0,0],[ax.get_ylim()[0],ax.get_ylim()[1]], color = 'w')
            plt.legend(loc = 'upper right', bbox_to_anchor=(1.1, 1.1),prop={'size':12})
            plt.text(0.1, 0.9,f, ha='center', va='center', transform=ax.transAxes, fontsize=24)
    
        ax.set_ylim([np.min(yscale),np.max(yscale)])
    canvas.print_figure(fig_filename)
```

```python
pth = '/Users/david_conant/Dropbox/Vowels/EC41/'
blocks = [19,20,22,27]
tokens = ['HAWED','HEED','WHOOD','HAD','HUD','HEAD','HEARD','HID','HOOD']
window = np.array([-1.5,2])
[K,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,dtype= 'kin',align_window = window)
Ks = [K[d] for d in tokens]
stops = [stop_times[d] for d in tokens]
starts = [start_times[d] for d in tokens]
Ks = process_kin(Ks,stops,starts,window, norm_length=False)
feats = ['Front Tongue X','Mid Tongue X', 'Back Tongue X','Front Tongue Y','Mid Tongue Y',
         'Back Tongue Y','Lip Opening','Lip Width','Jaw X','JawY']
labels = tokens
fig_filename = '/Users/david_conant/Dropbox/NeuroSpeech/Lab Meetings/Vowels1/EC41_Acous_hvdGS.png'
gestural_score(Ks,feats,labels,fig_filename,window)
```

```python
pth = '/Users/david_conant/Dropbox/Vowels/EC41/'
blocks = [19,20,22,27]
tokens = ['AAA','IYY','UWW','AEE','AHH','EHH','ERR','IHH','UHH']
window = np.array([-1,2])
[K,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,dtype='kin',align_window = window)
Ks = [K[d] for d in tokens]
stops = [stop_times[d] for d in tokens]
starts = [start_times[d] for d in tokens]
Ks = process_kin(Ks,stops,starts,window,numTongue=90)
Kall = np.concatenate(Ks,axis=2)
notes = []
for ind,n in enumerate(tokens):
    notes.append([n]*Ks[ind].shape[2])
notes = np.concatenate(notes,axis=0) 
feats = ['Front Tongue X','Mid Tongue X', 'Back Tongue X','Front Tongue Y','Mid Tongue Y',
         'Back Tongue Y','Lip Opening','Lip Width','Jaw X','JawY']
labels = tokens
#Timecourse
fig_filename = '/Users/david_conant/Dropbox/NeuroSpeech/Lab Meetings/Vowels1/EC41_Acous_GS.png'
gestural_score(Ks,feats,labels,fig_filename,window)


#LDA
mid = np.arange(37,42)
X = pd.DataFrame(np.nanmean(Kall[:,mid,:],axis=1).transpose())
X = X.fillna(X.median(),inplace=True)
clf = LDA()
Kproj = clf.fit_transform(X,notes)

sns.set_palette('Set1',n_colors=9)

sns.set_context('talk')
sns.set_style('dark')
fig, axes = plt.subplots(figsize=(10,10))
canvas = FigureCanvas(fig)
for cnt,tok in enumerate(tokens):
    thisTok = tok == notes
    k = Kproj[thisTok,:]
    ax = ppl.scatter(k[:,0],k[:,1], alpha = 1, s=60)
plt.xlabel('LD1')
plt.ylabel('LD2')
ppl.legend(labels, bbox_to_anchor=(1, 1))
#axes.set_ylim([1000,3100])
#axes.set_xlim([400,900])
fig_filename = '/Users/david_conant/Dropbox/NeuroSpeech/Lab Meetings/Vowels1/EC34_Acous_F1F2.png'
canvas.print_figure(fig_filename)
```
