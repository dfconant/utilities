```python
def Ds2Xy(Es,Ds,anat,dtype,vis=1,ncomp=10,erange=np.arange(366,370),frange=np.arange(105,115),krange=np.arange(37,40),elects=range(256)):
    #Feature labels
    forms = ['F0','F1','F2','F3','F4']
    kins = ['tongue1x','tongue2x','tongue3x','tongue1y','tongue2y','tongue3y','lipOpen', 'lipWide', 'jawx', 'jawy']
    
    #Anatomy
    #Formants
    if dtype is 'formants':
        Fall = np.concatenate(Fs,axis=2)
        dat = dict(zip(forms,np.nanmean(Fall[:,frange,:],axis = 1)))
        Fmid = pd.DataFrame.from_dict(dat)
        Fmid = Fmid.fillna(Fmid.median(),inplace=True)
        F2_1 = pd.Series(Fmid['F2']/Fmid['F1'],name = 'F2/F1')
        y = pd.concat((Fmid,F2_1),axis=1)
        y.apply(np.log10,axis = 0)
        y = (y - y.mean(axis = 0))/y.std(ddof=0)
        yrnd = shuffle_df(y,10,axis=0)
        y = y.values
        yrnd = yrnd.values
        #clf = PCA(n_components = 3) 
        #y = clf.fit_transform(y)
        #yrnd = clf.fit_transform(yrnd)
    #Kinematics
    elif dtype is 'kinematics':
        Kall = np.concatenate(Ks,axis=2)
        dat = dict(zip(kins,np.nanmean(Kall[:,krange,:],axis = 1)))
        Kmid = pd.DataFrame.from_dict(dat)
        Kmid = Kmid.fillna(Kmid.median(),inplace=True)
        y = Kmid
        y = (y - y.mean(axis = 0))/y.std(ddof=0)
        yrnd = shuffle_df(y,10,axis=0)
        y = y.values
        yrnd = yrnd.values
        #clf = PCA(n_components = 5) 
        #y = clf.fit_transform(y)
        #yrnd = clf.fit_transform(yrnd)
    #HG    
    Eall = np.concatenate(Es,axis=2)
    Eall = Eall[elects,:,:]
    X = np.nanmean(Eall[:,erange,:],axis = 1).transpose()
    X = pd.DataFrame(X)
    X = X.fillna(X.median(),inplace=True)
    X = X.dropna(axis =1,how='any')
    clf = PCA(n_components = ncomp) 
    clf.fit(X)
    var_ex = clf.explained_variance_ratio_
    plot(cumsum(var_ex))
    X = clf.transform(X)
    
    
    if vis:
        X = pd.DataFrame(X)
        y = pd.DataFrame(y)
        #X with X
        sns.pairplot(X,kind="reg")
        #y with y
        sns.pairplot(y,kind="reg")
        #X with y
        fig, axes = plt.subplots(figsize=(18,11))
        axes.axis('off')
        canvas = FigureCanvas(fig)
        c = 1
        for d1 in range(X.shape[1]):
            for d2 in range(y.shape[1]):
                ax = fig.add_subplot(y.shape[1],X.shape[1],c)
                #ppl.scatter(X[d1], y[d2])
                sns.regplot(X[d1], y[d2],ax=ax)
                c += 1
                
        X = X.values
        y = y.values
    return X,y,yrnd
```
