Percent variance explained is calculated as the r2 for each feature for each vowel, derived from a model traind on all vowels. These graphs are an attempt to visualize what part of the kinematics fall into 'acoustically dependent' and 'acoustically independent' categories. Percent variance explained here is equivalent to percent variance remaining in the 'acoustically dependent' component of the kinematics. 
I think there are several encouraging this about this plot. First, for most vowels, there is clear structure where only a few features are well predicted, and the rest are near zero. Second, although there are some unexpected datapoints, by and large I think these make sense. For example, front tongue is best predicted in IYY, but not for UWW. 
For the features that don't line up with what we expect, I think that's likely a reflection of the imperfection of the parameterization/model. Comparing across other parameterizations and subjects will give us a better picture.
