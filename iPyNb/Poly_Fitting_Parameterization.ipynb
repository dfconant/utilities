```python
#Intitialize
%load_ext autoreload
%autoreload 2
%matplotlib inline
import scipy as sp
import numpy as np
import pandas as pd
import prettyplotlib as ppl
from BDutils import ProgressBar
import seaborn as sns
from ECoGstats import makeD_multi, process_kin, smooth_formants, Ridge_MLR, Ds2Xy, procrustes_tongue
from ecog_viz import LDA_proj_scatter, gestural_score, separability_timecourse,feature_correlations,vowel_svm,R2viz,plotkhat
from __future__ import division
from matplotlib.backends.backend_agg import (FigureCanvasAgg as FigureCanvas)

sns.set_palette('Set1')
sns.set_context('talk')
```

```python
subj = 'EC41'
blocks = [19,20,22,27]
pth = '/Users/david_conant/Dropbox/Vowels/'+subj + '/'
tokens = ['AAA','AEE','AHH','EHH','ERR','IHH','IYY','UHH','UWW']
kfeats = ['c3','c2','c1','c0','Lip Opening','Lip Width','Jaw X','Jaw Y']
window = np.array([-1,2])
[K,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,dtype= 'kin',align_window = window)
Ks = [K[d].copy() for d in tokens]
#Ks,meanTng,Z,d = procrustes_tongue(Ks[:],scale=True,viz=False)
notes = []
for ind,n in enumerate(tokens):
    notes.append([n]*Ks[ind].shape[2])
notes = np.concatenate(notes,axis=0) 

stops = [stop_times[d] for d in tokens]
starts = [start_times[d] for d in tokens]

#Ks,meanTng,Z,d = procrustes_tongue(Ks[:],scale=True,viz=False)
Kt = process_kin(Ks[:],stops,starts,window, norm_length=False,velocity = False,wrt_start=True,poly_fit=3)
```

###How is it derived?

I fit a n-order polynomial to the tongue contour, then use the coefficients of the function as the features that describe the tongue. A 3rd order tends to give me the best performance

e.g. if a given contour is best fit by the equation $3x^3 + x^2 + 8x + 12$ then the coefficients for that tongue shape would be 3,1,8,and 12

This could give us a low-dimensional way of representing the position and, importantly, the shape of the tongue. This idea has been explored before in one of Stone's papers.

###Artic to Acoustic Mapping

```python
#Acoustic Mapping
elects = [21,25,35,37,43,53,55,58,59,67,72,73,86,90,102]
pth = '/Users/david_conant/Dropbox/Vowels/'+subj + '/'
window = np.array([-1,2])
ffeats = ['F0','F1','F2','F3','F4','F2/F1']

[F,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,align_window = window, dtype='form')
Fs = [F[d] for d in tokens]
Fs = smooth_formants(Fs,starts,stops,window)

CV = 'LOO'
boot = 1
ncomp = 8
nTP = 1
alphas = np.logspace(-2,3,20)
Fr2 = np.empty((Fs[0].shape[0]+1,nTP,len(alphas)))
Fr2null = np.empty((Fs[0].shape[0]+1,nTP,len(alphas)))
prog = ProgressBar(nTP)
prog.animate(0)
for indt,t in enumerate(np.linspace(0,396,nTP)):
    erange = np.arange(t,t+4)   #Try playing with window size

    #Organize data into X and y
    X,y,yrnd = Ds2Xy(Kt,Fs,dt1='kin',dt2='form',ncomp = ncomp,erange = erange.astype(int),elects = elects,frange=np.arange(105,115),krange=np.arange(37,42))

    #Ridge regressionnan
    for inda,a in enumerate(alphas):
        Fr2[:,indt,inda],yhat,score = Ridge_MLR(X, y, alpha = a, CV = CV, boot = boot)
        Fr2null[:,indt,inda],yhat,score = Ridge_MLR(X, yrnd, alpha = a, CV = CV, boot = boot)
    prog.animate(indt+1)
R2viz(Fr2,Fr2null,ffeats,b=False)
```

The coefficients are able to predict the acoustics fairly well, and in the pattern we would expect...

###Acoustic to Kinematic mapping/decomposition

```python
#Kinematic Mapping/Decomposition
elects = [21,25,35,37,43,53,55,58,59,67,72,73,86,90,102]
pth = '/Users/david_conant/Dropbox/Vowels/'+subj + '/'
window = np.array([-1,2])

[F,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,align_window = window, dtype='form')
Fs = [F[d] for d in tokens]
Fs = smooth_formants(Fs,starts,stops,window)

CV = 'LOO'
boot = 1
ncomp = 8
nTP = 1
alphas = np.logspace(-3,3,20)
Kr2 = np.empty((Kt[0].shape[0],nTP,len(alphas)))
Kr2null = np.empty((Kt[0].shape[0],nTP,len(alphas)))

prog = ProgressBar(nTP)
prog.animate(0)
for indt,t in enumerate(np.linspace(0,396,nTP)):
    erange = np.arange(t,t+4)   #Try playing with window size

    #Organize data into X and y
    X,y,yrnd = Ds2Xy(Fs,Kt,dt1='form',dt2='kin',ncomp = ncomp,erange = erange.astype(int),elects = elects,frange=np.arange(105,115),krange=np.arange(37,42))
    if indt is 0:
        Korig = np.empty((y.shape[0],y.shape[1],nTP,len(alphas)))
        Khat = np.empty((y.shape[0],y.shape[1],nTP,len(alphas)))
        Ktilde = np.empty((y.shape[0],y.shape[1],nTP,len(alphas)))
    #Ridge regressionnan
    for inda,a in enumerate(alphas):
        Kr2[:,indt,inda],yhat,score = Ridge_MLR(X, y, alpha = a, CV = CV, boot = boot)
        Korig[:,:,indt,inda] = squeeze(y)
        Khat[:,:,indt,inda] = squeeze(yhat)
        Ktilde[:,:,indt,inda] = y - squeeze(yhat)
        Kr2null[:,indt,inda],yhat,score = Ridge_MLR(X, yrnd, alpha = a, CV = CV, boot = boot)
    prog.animate(indt+1)
R2viz(Kr2,Kr2null,kfeats,b=False)
```

But in general they are very poorly predicted by the acoustics. To me, this suggests that the coefficients when combined (as in LDA, or PCA, or a linear regression) contain information fairly similar to earlier parameterizations, but separately, they aren't very well correlated to the acoustics or prior 

```python
plotkhat(Korig,Khat,kfeats,notes)
```

This I find fascinating. Although across all vowels, the coefficients are poorly predicted, for some individual vowels they are the best predicted feature (e.g. AHH,EHH). Furthermore, some of vowels appear to have structure across the coefficients (e.g. EHH's monotonically decreasing from high to low). To me, this suggests that there is something to interpret in the values of each coefficient (ie. there must be a reason by the cubic coefficient is better predicted than the square, etc). ###vSMC Decoding

```python
#Cross all subjects
#Load data

CV = 'split'
boot = 10
ncomp = 8
nTP = 200
alphas = np.logspace(-2,8,20)

window = np.array([-1, 2])
[E,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,align_window = window, dtype='HG')
Es = [E[d] for d in tokens]
elects = np.concatenate((np.where(anat=='preCG')[0],np.where(anat=='postCG')[0]))

Kr2 = np.empty((Kt[0].shape[0],boot,nTP,len(alphas)))
Kr2null = np.empty((Kt[0].shape[0],boot,nTP,len(alphas)))

prog = ProgressBar(nTP)
for indt,t in enumerate(np.linspace(200,600,nTP)):
    erange = np.arange(t,t+4)   #Try playing with window size

    #Organize data into X and y
    X,y,yrnd = Ds2Xy(Es,Kt,dt1='HG',dt2='kin',ncomp = ncomp,erange = erange.astype(int),elects = elects,frange=np.arange(105,115),krange=np.arange(37,42))

    #Ridge regression
    for inda,a in enumerate(alphas):
        Kr2[:,:,indt,inda],yhat,score = Ridge_MLR(X, y, alpha = a, CV = CV, boot = boot)
        Kr2null[:,:,indt,inda],yhat,score = Ridge_MLR(X, yrnd, alpha = a, CV = CV, boot = boot)
    prog.animate(indt)
R2viz(np.nanmean(Kr2,axis=1),np.nanmean(Kr2null,axis=1),kfeats,b=False)
```

While some of the prior analyses were interesting, this is a bit of a nail in the coffin for this parameterization. The coefficients are very poorly predicted by vSMC activity, comparable to the condition in which I shuffled the observations. So even if there is something interesting to be said about polynomial parameterization with respect to the acoustics, I think it's pretty clear that the brain does not represent movements in these terms. I don't plan to pursue it further for the time being.

###Conclusions


>When combined, the coefficients of a polynomial-fit of a contour relate to the vowel acoustics similarly to point-based parameterizations

>The dimensions of greatest variance seem to be very different than those of greatest discriminatory power (ie the organization of the PCs seem to be very different than those of the LDs)

>Separately, the coefficients are poorly predicted by the acoustics, altough this seems highly vowel dependent

>The coefficients are very poorly predicted by vSMC HG, suggesting that this is not the way that vSMC is organizing vowel movements. In other words, the brain doesn't do algebra.
