```python
pth = '/Users/david_conant/Documents/EC41/'
blocks = [19,20,22,27]
tokens = ['AAA','AEE','AHH','EHH','ERR','IHH','IYY','UHH','UWW']
window = np.array([-1, 4])
[F,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,align_window = window, dtype='form')
Fs = [F[d] for d in tokens]
stops = [stop_times[d] for d in tokens]
starts = [start_times[d] for d in tokens]
```

```python
Ds = ds[:]
form_sm = [None]*len(Ds)
t_length = []
for iD,D in enumerate(Ds):
    form_sm[iD] = np.empty(D.shape)
    for f in range(D.shape[0]):
        F = D[f,:,:]  
        for c in range(F.shape[1]):
            col = F[:,c]
            t_length.append(np.round((stops[iD][c] - starts[iD][c])*100))
            start = abs(window[0])*100
            end = start+t_length[-1]
            speak = np.zeros(col.shape,dtype = bool)
            speak[start:end] = True
            col[np.logical_and(col == 0,speak)] = np.nan
            col = savgol_filter(col,11,3)
            col[~speak] = 0
            col = pd.Series(col)
            col.interpolate(method = 'linear')
            F[:,c] = col.values 
        form_sm[iD][f,:,:] = F
med_length = np.round(np.median(t_length)) 
cnt = 0
for iD,D in enumerate(form_sm):
    for f in range(D.shape[0]):
        F = D[f,:,:] 
        for c in range(F.shape[1]):
            col = F[:,c]
            change_length = med_length - t_length[cnt]
            start = abs(window[0])*100
            end = start+t_length[cnt]
            newspeak = resample(col[start:end],med_length)
            nbef = start
            naft = col.shape[0] - end - change_length
            resamped = np.concatenate((np.zeros((nbef)),newspeak,np.zeros((naft))))
            F[:,c] = resamped
            cnt = cnt+1
        form_sm[iD][f,:,:] = F
```

```python
def gestural_score(Fs,feats,labels):

    co = ppl.brewer2mpl.get_map('Set1','Qualitative',9).mpl_colors
    align_window = window
    tt = np.linspace(align_window[0],align_window[1],Fs[0].shape[1])
    sns.set_context('talk')
    sns.set_style('dark')
    fig, axes = plt.subplots(figsize=(12,40))
    axes.axis('off')
    canvas = FigureCanvas(fig)

    for cnt,f in enumerate(feats):
        yscale = np.empty((len(Fs),2))
        ax = fig.add_subplot(len(feats),1,cnt+1)
        for iD,D in enumerate(Fs):
            mu = savgol_filter(np.nanmean(D[f,:,:],axis=1),5,3)
            ste = savgol_filter(np.nanstd(D[f,:,:],axis=1)/np.sqrt(D.shape[2]),5,3)
            yscale[iD,:] = np.array([np.nanmin(mu[50:75]-10*ste[50:75]), np.nanmax(mu[50:75]+10*ste[50:75])])
            plt.plot(tt,mu,color = co[iD],label=labels[iD])
            plt.fill_between(tt,mu-ste, mu+ste,color = co[iD],alpha=0.5)
            plt.plot([0,0],[ax.get_ylim()[0],ax.get_ylim()[1]], color = 'w')
            plt.legend(loc = 'upper right', bbox_to_anchor=(1.1, 1.1),prop={'size':12})
            plt.text(0.1, 0.9,'F' + str(f), ha='center', va='center', transform=ax.transAxes, fontsize=24)

        ax.set_ylim([np.min(yscale),np.max(yscale)])
```

```python
pth = '/Users/david_conant/Documents/EC41/'
blocks = [19,20,22,27]
tokens = ['AAA','AEE','AHH','EHH','ERR','IHH','IYY','UHH','UWW']
window = np.array([-.25, 2])
[F,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,align_window = window, dtype='form')
Fs = [F[d] for d in tokens]
stops = [stop_times[d] for d in tokens]
starts = [start_times[d] for d in tokens]
form_sm = smooth_formants(Fs,starts,stops,window)
feats = range(5)
labels = tokens
gestural_score(Fs,feats,labels)
```
