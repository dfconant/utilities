```python
# Create dataframes

pth = '/Users/david_conant/Dropbox/Vowels/EC41/'
blocks = [19,20,22,27]
tokens = ['AAA','IYY','UWW','AEE','AHH','EHH','ERR','IHH','UHH']
kfeats = ['FrontTongueX','MidTongueX', 'BackTongueX','FrontTongueY','MidTongueY',
         'BackTongueY','LipOpening','LipWidth','JawX','JawY']
lag = 0
fs = 100


#Kinematics
window = np.array([-0.5,0.5])
[K,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,dtype='kin',align_window = window)
Ks = [K[d] for d in tokens]
stops = [stop_times[d] for d in tokens]
starts = [start_times[d] for d in tokens]
Ks,meanTng,Z,d = procrustes_tongue(Ks[:],scale=True,viz=True)
Ks = process_kin(Ks,stops,starts,window,numTongue=200,wrt_start=False,norm_length=False)
Kall = np.concatenate(Ks,axis=2)
Kall = resample_array(Kall[:200,:,:],[200,np.diff(window)*fs,Kall.shape[2]])
```

```python
kfeats = np.hstack((seq_labels('tx',100),seq_labels('ty',100)))
Kinematics = np.reshape(Kall,(Kall.shape[0],-1),order='F').transpose()
Kinematics = pd.DataFrame(Kinematics,columns = kfeats)
#Kinematics.fillna(Kinematics.median(),inplace=True)
Kinematics.dropna(axis=0,how = 'all',inplace=True)
meantng = procrustes_mean(np.swapaxes(np.dstack((Kinematics.iloc[:,:100],Kinematics.iloc[:,100:])).transpose(),0,1),viz=True)
#Kinematics = pd.Series(dict(enumerate(Eall.transpose())))
```

```python
pca = PCA(n_components = 10)
pca.fit(Kinematics)
eigVect = pca.components_
scores = pca.explained_variance_ratio_
```
