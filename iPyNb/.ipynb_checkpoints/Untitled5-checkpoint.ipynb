```python
#Intitialize
%load_ext autoreload
%autoreload 2
%matplotlib inline
import scipy as sp
import numpy as np
from sklearn.cross_validation import train_test_split, LeaveOneOut, LeaveOneLabelOut, ShuffleSplit
import seaborn as sns
from ECoGstats import makeD_multi, smooth_formants, shuffle_df
from sklearn.decomposition import PCA
import pandas as pd
import prettyplotlib as ppl
import pdb
from BDutils import ProgressBar
from scipy.signal import savgol_filter
import matplotlib.gridspec as gridspec
import scipy.stats as stats
from numpy.linalg import lstsq
from scipy import stats
from __future__ import division
```

```python
def np_MLR_CV(X , y,  boot = 100, CV = 'split'):

    #Preallocate
    r2 = np.empty((y.shape[1],boot))
    slope = np.empty((y.shape[1],X.shape[1]+1,boot))
    #Add column of ones
    X = np.hstack([np.ones((X.shape[0],1)),X])

    #Bootstrapped, cross validated linear regression
    yhat = np.empty((y.shape))
    for b in range(boot):
        if CV is 'split':
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

            for c in range(y.shape[1]):
                #slope[c,:,b], intercept, r_value, p_value, std_err = stats.linregress(X_train,y_train[:,c])
                slope[c,:,b] = lstsq(X_train,y_train[:,c])[0]
                intercept = slope[c,0,b] 

                yhat = np.dot(slope[c,1:,b],X_test[:,1:].transpose()) + intercept
                r = stats.pearsonr(yhat,y_test[:,c])[0]
                r2[c,b] = r**2
        elif CV is 'LOO':
            loo = LeaveOneOut(X.shape[0])
            for c in range(y.shape[1]):

                for train_index, test_index in loo:
                    slope[c,:,b] = lstsq(X[train_index,:],y[train_index,c])[0]
                    intercept = slope[c,0,b] 
                    yhat[test_index,c] = np.dot(slope[c,1:,b],X[test_index,1:].transpose()) + intercept

                r = stats.pearsonr(yhat[:,c],y[:,c])[0]
                r2[c,b] = r**2
        elif CV is 'none':
            for c in range(y.shape[1]):
                #slope[c,:,b], intercept, r_value, p_value, std_err = stats.linregress(X_train,y_train[:,c])
                slope[c,:,b] = lstsq(X,y[:,c])[0]
                intercept = slope[c,0,b] 

                yhat[:,c] = np.dot(slope[c,1:,b],X[:,1:].transpose()) + intercept
                r = stats.pearsonr(yhat[:,c],y[:,c])[0]
                r2[c,b] = r**2
        elif CV is 'LOLO':
            trialLabel = np.tile(np.arange(y.shape[0]/3),3)
            lol = LeaveOneLabelOut(trialLabel)
            for c in range(y.shape[1]):

                for train_index, test_index in lol:
                    slope[c,:,b] = lstsq(X[train_index,:],y[train_index,c])[0]
                    intercept = slope[c,0,b] 
                    yhat[test_index,c] = np.dot(slope[c,1:,b],X[test_index,1:].transpose()) + intercept

                r = stats.pearsonr(yhat[:,c],y[:,c])[0]
                r2[c,b] = r**2
        elif CV is 'LPLO':
            numTrials = int(y.shape[0]/3)
            rs = ShuffleSplit(numTrials, n_iter=1,test_size=.2)
            for c in range(y.shape[1]):
                for train_index, test_index in rs:
                    test_trials = np.concatenate((test_index,test_index+numTrials,test_index+2*numTrials))
                    train_trials = np.concatenate((train_index,train_index+numTrials,train_index+2*numTrials))

                    slope[c,:,b] = lstsq(X[train_trials,:],y[train_trials,c])[0]
                    intercept = slope[c,0,b] 
                    yhat = np.dot(slope[c,1:,b],X[test_trials,1:].transpose()) + intercept
                    r = stats.pearsonr(yhat,y[test_trials,c])[0]
                    r2[c,b] = r**2
    return r2, yhat
```

```python
def MLR_static(Es,Fs,anat,boot, notes, ncomp = 10, CV = 'split'):
    #Feature labels
    forms = ['F0','F1','F2','F3','F4']
    chans = np.arange(1,256)
    #Anatomy
    vSMC = np.concatenate((np.where(anat=='preCG')[0],np.where(anat=='postCG')[0]))

    #Formants
    mid = np.arange(105,115)
    Fall = np.concatenate(Fs,axis=2)
    dat = dict(zip(forms,np.nanmean(Fall[:,mid,:],axis = 1)))
    Fmid = pd.DataFrame.from_dict(dat)
    Fmid = Fmid.fillna(Fmid.median(),inplace=True)
    F2_1 = pd.Series(Fmid['F2']/Fmid['F1'],name = 'F2/F1')
    y = pd.concat((Fmid,F2_1),axis=1)
    ##Adjacent Points
    tmp = pd.concat((y,y),axis = 0,ignore_index=True)
    y = pd.concat((tmp,y),axis = 0,ignore_index=True)
    ##
    y.apply(np.log10,axis = 0)
    y = (y - y.mean())/y.std(ddof=0)
    yrnd = shuffle_df(y,10,axis=0)
    y = y.values
    yrnd = yrnd.values
    
    erange = np.arange(368,370)  #Static window of -100:-60 ms
    Eall = np.concatenate(Es,axis=2)
    Eall = Eall[vSMC,:,:]
    X = np.nanmean(Eall[:,erange,:],axis = 1).transpose()
    ##Adjacent Points
    X1 = np.nanmean(Eall[:,erange-2,:],axis = 1).transpose()
    X3 = np.nanmean(Eall[:,erange+2,:],axis = 1).transpose()
    X = np.concatenate((X,X1,X3),axis = 0)
    ##
    X = pd.DataFrame(X)
    X = X.fillna(X.median(),inplace=True)
    X = X.dropna(axis =1,how='any')
    clf = PCA(n_components = ncomp) 
    clf.fit(X)
    X = clf.transform(X)
    ## Z-score inputs
    X = sp.stats.mstats.zscore(X,axis = 0)
    ##
    r2,yhat = np_MLR_CV(X,y,boot,CV)
    r2null,tmp = np_MLR_CV(X,yrnd,boot,CV)
    return r2, r2null, yhat, y
```
