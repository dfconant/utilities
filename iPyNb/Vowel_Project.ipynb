###By Vowel Type

First, for the sake of clarity, starting with just the corner vowels

```python
#Slice data
vow = ['HAWED','HEED','WHOOD']
annot1 = e.annotations(1)
Ds = [];
for v in vow:
    filt = annot1.description_contains_filter(v)
    a = annot1[filt]
    anno_data, anno_types, anno_descriptions, annos = a.data(stack=True)
    Ds.append(anno_data[:,100:400,:])
```

```python
ax = ev.plotERPs(Ds,align_window = np.array([-0.5,1]),anatomy=np.array(e.dims(axis = 'space', get_hdf = False)[2]['dataset']),labels = vow)
```

And for all vowels

```python
#Slice data
vow = np.unique(np.array(e.dims(axis = 'channels', get_hdf = False)[1]['dataset']))
annot1 = e.annotations(1)
Ds = [];
for v in vow:
    filt = annot1.description_contains_filter(v)
    a = annot1[filt]
    anno_data, anno_types, anno_descriptions, annos = a.data(stack=True)
    Ds.append(anno_data[:,100:400,:])
```
