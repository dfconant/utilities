```python
def get_data(subj,blocks):
    pth = '/Users/david_conant/Dropbox/Vowels/'+subj + '/'

    tokens = ['AAA','IYY','UWW','AEE','AHH','EHH','ERR','IHH','UHH']
    #tokens = ['HAWED','HEED','WHOOD','HAD','HUD','HEAD','HEARD','HID','HOOD']
    kfeats = ['kFrontTongueX','kFMidTongueX','kMidTongueX','kBMidTongueX','kBackTongueX','kFrontTongueY','kFMidTongueY','kMidTongueY',
              'kBMidTongueY','kBackTongueY','kLipOpening','kLipWidth','kJawX','kJawY']
    efeats = seq_labels('e',256)
    fs = 100
    #Load in Dmatrices

    #Electrodes
    window = np.array([-1,1])   #TODO: try wider window for more timepoints
    [E,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,align_window = window, dtype='HG')
    Es = [E[d] for d in tokens]
    vSMC = np.concatenate((np.where(anat=='preCG')[0],np.where(anat=='postCG')[0]),87)
    vSMC_logic = np.zeros(256,dtype=bool)
    vSMC_logic[vSMC] = True
    Eall = np.concatenate(Es,axis=2)
    Eall = resample_array(Eall,[Eall.shape[0],np.diff(window)*fs,Eall.shape[2]])

    #Kinematics
    #window = np.array([0.25,0.75])
    [K,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,dtype='kin',align_window = window)
    Ks = [K[d] for d in tokens]
    stops = [stop_times[d] for d in tokens]
    starts = [start_times[d] for d in tokens]
    Ks = process_kin(Ks,stops,starts,window,numTongue=10,wrt_start=True,norm_length=False)
    Kall = np.concatenate(Ks,axis=2)
    Kall = resample_array(Kall,[Kall.shape[0],np.diff(window)*fs,Kall.shape[2]])
    
    return Kall,Eall,anat
```

```python
def do_correlations(Kall,Eall,anat,decorr = False):
#################################
    #Do correlations
    fs = 100
    vSMC = np.concatenate((np.where(anat=='preCG')[0],np.where(anat=='postCG')[0]))
    #vSMC = np.where((anat!='STG') & (anat!='MTG') & (anat!='MTG'))[0]
    vSMC_logic = np.zeros(256,dtype=bool)
    vSMC_logic[vSMC] = True
    
    tokens = ['AAA','IYY','UWW','AEE','AHH','EHH','ERR','IHH','UHH','HAWED','HEED',
              'WHOOD','HAD','HUD','HEAD','HEARD','HID','HOOD']
    #tokens = ['HAWED','HEED','WHOOD','HAD','HUD','HEAD','HEARD','HID','HOOD']
    kfeats = ['kFrontTongueX','kFMidTongueX','kMidTongueX','kBMidTongueX','kBackTongueX','kFrontTongueY',
              'kFMidTongueY','kMidTongueY','kBMidTongueY','kBackTongueY','kLipOpening','kLipWidth','kJawX','kJawY']
    efeats = seq_labels('e',256)

    lags = np.linspace(-1,.5,40)
    Rs = np.empty((14,256,len(lags)))
    Rp = np.empty((14,256,len(lags)))
    ps = np.empty((14,256,len(lags)))
    pp = np.empty((14,256,len(lags)))
    prog = ProgressBar(len(lags))
    groups = [range(10),range(10,12),range(12,14)]
    win_size=0.5
    win_start = 1
    Rpc = []
    for ilag, lag in enumerate(lags):

        #Lag by removing data
        t_unlagged = np.arange(75,125)
        t_lagged = np.arange(np.floor((win_start+lag)*fs),np.floor((win_start+win_size)*fs+lag*fs)).astype(int)
        Klagged = Kall[:,t_unlagged,:]
        Elagged = Eall[:,t_lagged,:]   #behavior-leading lag

        #Reshape across trials
        Kinematics = np.reshape(Klagged,(Klagged.shape[0],-1),order='F').transpose()
        Kinematics = pd.DataFrame(Kinematics,columns = kfeats)
        #Kinematics.fillna(Kinematics.median(),inplace=True)
        #Kinematics = pd.Series(dict(enumerate(Eall.transpose())))

        HG = np.reshape(Elagged,(Elagged.shape[0],-1),order='F').transpose()
        HG = pd.DataFrame(HG,columns = efeats)
        vsmcHG = HG#.iloc[:,vSMC]
        #vsmcHG.dropna(axis =1,how='any',inplace=True)
        data = pd.concat([vsmcHG,Kinematics],axis=1)
        data.dropna(axis=1,how='all',inplace=True)   #Drop bad features
        data.dropna(axis=0,how='all',inplace=True)   #Drop bad observations
        data.dropna(axis=0,how='any',inplace=True)   #Drop bad observations
        data.dropna(axis=1,how='any',inplace=True)   #Drop bad features
        good_efeats = data.filter(regex=r'^e\S', axis=1).columns.values
        good_kfeats = data.filter(regex=r'^k\S', axis=1).columns.values
        if decorr:
            hat,tilde = SemiPartialCorrelation(data[Kinematics.columns],groups)
            data[Kinematics.columns] = tilde
        #Rpc.append(partial_corr(data))
        #groups = [range(10),range(10,12),range(12,14)]
        #for g in groups:
            #Rpc[g] = 

        good_elects = []
        inner = Kinematics.shape[1]
        outer = vsmcHG.shape[1]
        for e in good_efeats:
            good_elects.append(int(e[1:]))
            for i,k in enumerate(good_kfeats):
                #Rs[i,good_elects[-1],ilag],ps[i,good_elects[-1],ilag] = sp.stats.spearmanr(data.loc[:,e],data.loc[:,k])
                Rp[i,good_elects[-1],ilag],pp[i,good_elects[-1],ilag] = sp.stats.pearsonr(data.loc[:,e],data.loc[:,k])
        prog.animate(ilag+1)

    #########################################
    #plot stuff
    good_vSMC_logic = np.zeros(256,dtype=bool)
    good_vSMC_logic[good_elects] = True
    Rp[:,~good_vSMC_logic,:] = 0
    pp[:,~good_vSMC_logic,:] = 1
    Rp[abs(Rp) > 1] = 0
    #Rs[abs(Rs) > 1] = 0
    pp[abs(pp) > 1 ]= 1
    #ps[abs(ps) > 1] = 1
    pp = abs(1-pp)
    lips = np.nanmean(np.square(Rp[10:12,:,0]),axis=0)
    tonguey = np.nanmean(np.square(Rp[5:10,:,0]),axis=0)
    tonguex = np.nanmean(np.square(Rp[0:5,:,0]),axis=0)
    tongueF = np.nanmean(np.square(Rp[[0,1,5,6],:,0]),axis=0)
    tongueB = np.nanmean(np.square(Rp[[3,4,8,9],:,0]),axis=0)
    jaw = np.nanmean(np.square(Rp[12:,:,0]),axis=0)

    plips = np.nanmean(np.square(pp[10:12,:,0]),axis=0)
    ptonguey = np.nanmean(np.square(pp[5:10,:,0]),axis=0)
    ptonguex = np.nanmean(np.square(pp[0:5,:,0]),axis=0)
    ptongueF = np.nanmean(np.square(pp[[0,1,5,6],:,0]),axis=0)
    ptongueB = np.nanmean(np.square(pp[[3,4,8,9],:,0]),axis=0)
    pjaw = np.nanmean(np.square(pp[12:,:,0]),axis=0)

    artics = np.vstack((tonguex,tonguey,lips,jaw)).transpose()
    #ID electrodes according to their correlation with articulators
        #according to max r values
    rpc = Rp[:,:,:26]
    max_elects = np.amax(abs(rpc),axis=2)
    shaped_max_elects = np.reshape(max_elects.copy(),(14,16,16))
    vSMC_shaped = np.rot90(np.reshape(good_vSMC_logic,(16,16)),k=3)
    for i in range(14):
        shaped_max_elects[i,:,:] = np.rot90(shaped_max_elects[i,:,:],k=3)

    

    #fig_filename = save_pth + subj + '_correlation_somatotopy.png'
    #savefig(fig_filename,facecolor='k', edgecolor='k',bbox_inches='tight')
    
    return shaped_max_elects,vSMC_shaped
```

```python
def plot_corrs(to_plot,elects,do_plot=False):
    
    out = np.empty((16,16,14))
    kfeats = ['kFrontTongueX','kFMidTongueX','kMidTongueX','kBMidTongueX','kBackTongueX','kFrontTongueY',
          'kFMidTongueY','kMidTongueY','kBMidTongueY','kBackTongueY','kLipOpening','kLipWidth','kJawX','kJawY']
    if do_plot:
        fig = plt.subplots(figsize=(20,60))
    for feat in range(14):
        plotted = to_plot[feat,:,:]
        
        #plt.pcolor(plotted,cmap='gray')
        #plt.colorbar()
        #plt.gca().set_aspect('equal')
        #g_filt = sp.ndimage.filters.gaussian_filter(plotted,0.6,order=0)
        #g_filt = sp.ndimage.morphology.grey_erosion(g_filt,size=(2,2))
        g_filt = sp.ndimage.morphology.grey_dilation(plotted,size=(2,2))
        g_filt = sp.ndimage.filters.gaussian_filter(g_filt,0.6,order=0)
        g_filt = sp.ndimage.morphology.grey_erosion(g_filt,size=(1,1))
        g_filt = sp.ndimage.filters.gaussian_filter(g_filt,0.6,order=0)

        g_filt[~elects] = 0
        #g_filt[g_filt<0.04]=0.04
        if do_plot:
            plt.subplot(7,2,feat+1)
            plt.pcolor(g_filt,cmap='gist_heat')
            plt.colorbar()
            plt.gca().set_aspect('equal')
            plt.title(kfeats[feat][1:])

        out[:,:,feat] = g_filt

    
    return out
```

```python
#EC92############
subj = 'EC92'
blocks = [35,36,37,45]
Kall,Eall,anat = get_data(subj,blocks)
to_plot,elects = do_correlations(Kall,Eall,anat,decorr=False)
vSMC = (anat=='preCG') | (anat=='postCG')
vSMC = np.rot90(np.reshape(vSMC,(16,16)),k=3)
vSMC92_weights = plot_corrs(to_plot,vSMC)


#EC41############
subj = 'EC41'
blocks = [19,20,22,27]
Kall,Eall,anat = get_data(subj,blocks)
to_plot,elects = do_correlations(Kall,Eall,anat,decorr=False)
vSMC = (anat=='preCG') | (anat=='postCG')
vSMC = np.rot90(np.reshape(vSMC,(16,16)),k=3)
vSMC41_weights = plot_corrs(to_plot,vSMC)
```

#Compare effects of decorrelation

#### One important caveat to my prior analyses was that the kinematic features were not independent, and thus overlapping results have ambiguous interpretations. As a starting place, I tried running the kinematic features through a semi-partial correlation before doing the correlations with the electrodes. This will yield the kinematic features that are linearly indpendent of the others, and thus shared correlations can (tentatively) be interpreted as a true overlapping representation.

```python
#EC92############
subj = 'EC92'
blocks = [35,36,37,45]
Kall,Eall,anat = get_data(subj,blocks)
to_plot,elects = do_correlations(Kall,Eall,anat,decorr=True)
vSMC = (anat=='preCG') | (anat=='postCG')
vSMC = np.rot90(np.reshape(vSMC,(16,16)),k=3)
vSMC92_weights_decorr = plot_corrs(to_plot,vSMC)


#EC41############
subj = 'EC41'
blocks = [19,20,22,27]
Kall,Eall,anat = get_data(subj,blocks)
to_plot,elects = do_correlations(Kall,Eall,anat,decorr=True)
vSMC = (anat=='preCG') | (anat=='postCG')
vSMC = np.rot90(np.reshape(vSMC,(16,16)),k=3)
vSMC41_weights_decorr = plot_corrs(to_plot,vSMC)
```

```python
groups = [range(10),range(10,12),range(12,14)]
labs = ['Tongue','Lips','Jaw']

for i,g in enumerate(groups):
    fig = plt.subplots(2,2,figsize=[20,15])
    plt.subplot(2,2,1)
    vSMC41 = np.nanmean(vSMC41_weights[:,:,g],axis=2)
    plt.pcolor(vSMC41,cmap='gist_heat')
    plt.colorbar()
    plt.gca().set_aspect('equal')
    plt.title(labs[i] + ' vSMC_EC41')
    
    plt.subplot(2,2,2)
    vSMC92 = np.nanmean(vSMC92_weights[:,:,g],axis=2)
    plt.pcolor(vSMC92,cmap='gist_heat')
    plt.colorbar()
    plt.gca().set_aspect('equal')
    plt.title(labs[i] + ' vSMC_EC92')
    
    plt.subplot(2,2,3)
    vSMC41_decorr = np.nanmean(vSMC41_weights_decorr[:,:,g],axis=2)
    plt.pcolor(vSMC41_decorr,cmap='gist_heat')
    plt.colorbar()
    plt.gca().set_aspect('equal')
    plt.title(labs[i] + ' vSMC_EC41_decorrelated')
    
    plt.subplot(2,2,4)
    vSMC92_decorr = np.nanmean(vSMC92_weights_decorr[:,:,g],axis=2)
    plt.pcolor(vSMC92_decorr,cmap='gist_heat')
    plt.colorbar()
    plt.gca().set_aspect('equal')
    plt.title(labs[i] + ' vSMC_EC92_decorrelated')
    
    plt.suptitle(labs[i],fontsize=20)
    plt.show()
```

#### Decorrelating the tongue doesn't change much, which is not surprising considering that very little of the variance of the tongue is shared with the lips or jaw. The jaw and lips change a bit, but not drastically. If anything, it appears to spread out slightly. To me, this suggests that the overlapping regions we see are real rather than the result of statistical dependence.

I debated showing this to you because I don't have a clear idea of what this means, but I think the possibility of linking vSMC activity to other areas could be compelling.
