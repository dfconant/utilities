**The primary systematic feature I see is that UW and IY are consistently among the best, if not the best, classified vowels: often with accuracies that are very respectable.** 

**Let's take a step back and think about what we've observed so far:**

    -When looking at only the corner vowel ERPs, we see many electrodes that seems tuned, especially to IY or UW.
    -However, when we fill in all the vowels they don't appear to 'fill in' the gaps in a graded fashion (as one would expect if there were tuning for the continuous features that determine vowel identity (e.g. front tongue height)). Instead, the only observable discrimination is with UW or IY.
    -Classification confusion matrices reiterate this conclusion, as the only consistently well predicted vowels are UW and IY.
    

We knew that in the behavioral measures the rest of the vowels 'fill in' the space between the corner vowels, so we naturally hypothesized that the rest of the vowels would do the same. However, it appears that this is not the case. In fact, the only outliers of the vowels in the neural data seem to be UW and IY. In other words, the vowels we selected to be most discriminable in previous studies are much more dramatically different than we anticipated.

As a quick (necessary but not sufficient) way of checking if this is plausable, what does the classification look like if we restrict to just the vowels used in prior analyses? (AA/IY/UW). We should be getting performance comparable to what Kris found.
