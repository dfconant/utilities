```python
#def formant_boats_decode_timecourse(Es,Fs,tokens,anat,alphas = np.tile(np.logspace(-10,10,50),(6,1)), ncomp = 50):
alphas = np.tile(np.logspace(-10,10,50),(6,1))
ncomp = 50
#Trial labels
notes = []
for ind,n in enumerate(tokens):
    notes.append([n]*Fs[ind].shape[2])
notes = np.concatenate(notes,axis=1)
#Feature labels
forms = ['F0','F1','F2','F3','F4']
chans = np.arange(1,256)
#Anatomy
vSMC = np.concatenate((np.where(anat=='preCG')[0],np.where(anat=='postCG')[0]))

#Formants = central 20% of each vowel utterance 
mid = np.arange(105,115)
Fall = np.concatenate(Fs,axis=2)
dat = dict(zip(forms,np.nanmean(Fall[:,mid,:],axis = 1)))
#dat['label'] = notes
Fmid = pd.DataFrame.from_dict(dat)
Fmid = Fmid.fillna(Fmid.median(),inplace=True)
F2_1 = pd.Series(Fmid['F2']/Fmid['F1'],name = 'F2/F1')
y = pd.concat((Fmid,F2_1),axis=1)
#y.apply(np.log10,axis = 0)
y = (y - y.mean())/y.std(ddof=0)


#Rolling average of vSMC HG   
centers = np.round(np.linspace(140,440,num = 75))     #10 ms window. From -650 to 100
half_width = 2
Ebins = [None]*centers.shape[0]
Eall = np.concatenate(Es,axis=2)
Eall = Eall[vSMC,:,:]



boot = 5
boat_boot = 200
r2 = np.empty((boot,len(Ebins),y.shape[1],alphas.shape[1]))
R2 = np.empty((len(Ebins),y.shape[1],alphas.shape[1]))
R2ste = np.empty((len(Ebins),y.shape[1],alphas.shape[1]))
prog = ProgressBar(len(centers))
for ind,cent in enumerate(centers.astype(int)):
    erange = np.arange(cent-half_width,cent+half_width)
    X = np.nanmean(Eall[:,erange,:],axis = 1).transpose()
    X = pd.DataFrame(X)
    X = X.fillna(X.median(),inplace=True)
    X = X.dropna(axis =1,how='any')

    #PCA
    clf = PCA(n_components = ncomp) 
    clf.fit(X)
    X = clf.transform(X)

    for inda,alph in enumerate(alphas.transpose()):
        



        #BoATS Regression

        #Find sig weigths
        Brnd = np.empty((y.shape[1],ncomp,boat_boot))
        Bobs = np.empty((y.shape[1],ncomp,boat_boot))
        sigB = np.empty((y.shape[1],ncomp))               
        for bb in range(boat_boot):

            yrnd = shuffle_df(y,10,axis=0)
            X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.2)
            yrnd_train, yrnd_test = train_test_split(
            yrnd, test_size=0.2)

            for c,f in enumerate(y.columns.values):

                clf = LinearRegression(fit_intercept = False, normalize = True, n_jobs=-1)   
                clf.fit(X_train,yrnd_train.loc[:,f])
                Brnd[c,:,bb] = clf.coef_

                clf = LinearRegression(fit_intercept = False, normalize = True, n_jobs=-1)   
                clf.fit(X_train,y_train.loc[:,f])
                Bobs[c,:,bb] = clf.coef_

        sigB = [None]*6
        for c,f in enumerate(y.columns.values): 
            sigB[c] = np.where(np.median(np.abs(Bobs[c,:,:]),axis = 1) 
                   > 2* np.nanstd(np.median(np.abs(Brnd[c,:,:]),axis = 1),axis=0))[0]
        for b in range(boot):
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
            for c,f in enumerate(y.columns.values):
                X_train_r = X_train[:,sigB[c]]
                X_test_r = X_test[:,sigB[c]]
                
                clf = LinearRegression(fit_intercept = False, normalize = True, n_jobs=-1)   
                clf.fit(X_train_r,y_train.loc[:,f])
                yhat = clf.predict(X_test_r)
            
                r = sp.stats.pearsonr(yhat,y_test.loc[:,f])[0]
                r2[b,ind,c] = r**2 

        R2[ind,:] = np.mean(r2[:,ind,:],axis = 0)
        R2ste[ind,:] = np.std(r2[:,ind,:],axis = 0)/boot
        prog.animate(ind)

#Best performing 
best_times = centers[np.argmax(R2,axis = 0)]

fig = plt.figure(figsize = (10,10))
gs1 = gridspec.GridSpec(10,10)
co = ppl.brewer2mpl.get_map('Dark2','Qualitative',6).mpl_colors
tt = np.linspace(-650,100,75)
forms = ['F0','F1','F2','F3','F4','F2/F1']
sns.set_context('talk')
ax1 = plt.subplot(gs1[6:10,1:10])
ax1.set_ylabel('R2')
ax1.set_xlabel('lag (ms)')
#ax1.set_title('-100 ms')
for ind,f in enumerate(forms):
    sig = savgol_filter(R2[:,ind],11,3)
    ste = savgol_filter(R2ste[:,ind],11,3)
    plt.plot(tt,sig,color = co[ind],label=f)
    plt.fill_between(tt,sig-ste, sig+ste,color = co[ind],alpha=0.5)
plt.legend(loc = 'upper right', bbox_to_anchor=(1.1, 1.1),prop={'size':12})

ax2 = plt.subplot(gs1[1:5,1:5])
plt.pcolor(R2[:,5,:])
plt.colorbar()
return R2,R2ste,r2,best_alphas,best_times
```
