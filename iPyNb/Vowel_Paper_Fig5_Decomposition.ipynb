##Linear prediction of kinematics from formants

```python
pth = '/Users/david_conant/Dropbox/Vowels/EC41/'
blocks = [19,20,22,27]
tokens = ['AAA','AEE','AHH','EHH','ERR','IHH','IYY','UHH','UWW']
#tokens = ['HAWED','HEED','WHOOD','HAD','HUD','HEAD','HEARD','HID','HOOD']
kfeats = ['kFrontTongueX','kFMidTongueX','kMidTongueX','kBMidTongueX','kBackTongueX','kFrontTongueY','kFMidTongueY','kMidTongueY',
          'kBMidTongueY','kBackTongueY','kLipOpening','kLipWidth','kJawX','kJawY']
efeats = seq_labels('e',256)
fs = 100

window = np.array([-1,2])   

[K,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,dtype='kin',align_window = window)
Ks = [K[d] for d in tokens]
stops = [stop_times[d] for d in tokens]
starts = [start_times[d] for d in tokens]
Ks = process_kin(Ks,stops,starts,window,numTongue=10,wrt_start=False,norm_length=False)

[F,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,align_window = window, dtype='form')
Fs = [F[d] for d in tokens]
Fs = smooth_formants(Fs,starts,stops,window)

CV = 'LOO'
boot = 1
ncomp = 8
nTP = 1
alphas = np.logspace(-3,3,20)
Kr2 = np.empty((Ks[0].shape[0],nTP,len(alphas)))
Kr2null = np.empty((Ks[0].shape[0],nTP,len(alphas)))

prog = ProgressBar(nTP)
prog.animate(0)
for indt,t in enumerate(np.linspace(0,396,nTP)):
    erange = np.arange(t,t+4)   

    #Organize data into X and y
    X,y,yrnd = Ds2Xy(Fs,Ks,dt1='form',dt2='kin',ncomp = ncomp,erange = erange.astype(int),elects = [],frange=np.arange(105,115),krange=np.arange(37,42))
    if indt is 0:
        Korig = np.empty((y.shape[0],y.shape[1],nTP,len(alphas)))
        Khat = np.empty((y.shape[0],y.shape[1],nTP,len(alphas)))
        Ktilde = np.empty((y.shape[0],y.shape[1],nTP,len(alphas)))
    #Ridge regression
    for inda,a in enumerate(alphas):
        Kr2[:,indt,inda],coefs,yhat,score = Ridge_MLR(X, y, alpha = a, CV = CV, boot = boot)
        Korig[:,:,indt,inda] = squeeze(y)
        Khat[:,:,indt,inda] = squeeze(yhat)
        Ktilde[:,:,indt,inda] = y - squeeze(yhat)
        Kr2null[:,indt,inda],coefs,yhat,score = Ridge_MLR(X, yrnd, alpha = a, CV = CV, boot = boot)
    prog.animate(indt+1)
#R2viz(Kr2,Kr2null,kfeats,b=False)

notes = []
for ind,n in enumerate(tokens):
    notes.append([n]*Ks[ind].shape[2])
notes = np.concatenate(notes,axis=0) 
plotkhat(Korig,Khat,kfeats,notes)
```

####Plotted here is the percent variance explained (R<sup>2</sup>) of the kinematics modeled from the acoustics, divided by vowel and kinematic feature. Put another way, this is the portion of the variance of the measured kineamtics that falls into acoustically relevant category (K<sup>hat</sup>). From this we can see 1)There is a lot of variability in R<sup>2</sup> across both vowels and kinematic feature. 2)The features that are high for a given vowel tend to be the features that are cannonically described as the critical features. For example, for /AA/ the front tongue frontness and mid tongue height are especially well predicted by the acosutics. This is promising as it means that our decomposition is identifying specific features (which is interpretable) rather than some strange combination of all features (which is difficult to interpret).
