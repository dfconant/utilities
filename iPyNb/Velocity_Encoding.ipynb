```python
#Intitialize
%load_ext autoreload
%autoreload 2
%matplotlib inline
import scipy as sp
import scipy.stats as stats
import numpy as np
from sklearn.cross_validation import train_test_split, LeaveOneOut, LeaveOneLabelOut, ShuffleSplit
from sklearn.decomposition import PCA
from sklearn.lda import LDA
from sklearn.linear_model import Lars, Ridge
import pandas as pd
import pdb
import prettyplotlib as ppl
from BDutils import ProgressBar, moving_average
from scipy.signal import savgol_filter
import seaborn as sns
import matplotlib.gridspec as gridspec
from ECoGstats import makeD_multi, process_kin, shuffle_df, smooth_formants, Ridge_MLR, Ds2Xy, smooth_derivative,ClusterSeparability
from ecog_viz import LDA_proj_scatter, gestural_score, separability_timecourse,feature_correlations,vowel_svm
from __future__ import division
from matplotlib.backends.backend_agg import (FigureCanvasAgg as FigureCanvas)
from scipy.stats import mstats
```

```python
subj = 'EC41'
blocks = [19,20,22,27]
pth = '/Users/david_conant/Dropbox/Vowels/'+subj + '/'
tokens = ['AAA','IYY','UWW','AEE','AHH','EHH','ERR','IHH','UHH']
window = np.array([-1,2])
[K,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,dtype= 'kin',align_window = window)
Ks = [K[d].copy() for d in tokens]
notes = []
for ind,n in enumerate(tokens):
    notes.append([n]*Ks[ind].shape[2])
notes = np.concatenate(notes,axis=0) 

stops = [stop_times[d] for d in tokens]
starts = [start_times[d] for d in tokens]
Ks = process_kin(Ks,stops,starts,window, norm_length=True,velocity = True,wrt_start=True)
Kall = np.concatenate(Ks,axis=2)
feats = ['Front Tongue X','Mid Tongue X', 'Back Tongue X','Front Tongue Y','Mid Tongue Y',
         'Back Tongue Y','Lip Opening','Lip Width','Jaw X','Jaw Y']
sns.set_palette('Set1')

#Gestural Score:
#fig_filename = save_pth + subj + '_Kin_hvdGS_legend.png'
gestural_score(Ks,feats,tokens,fig_filename =None,window=np.array([-1,2]))
```

```python
#Cluster Separability:
Ks = [K[d].copy() for d in tokens]
Ks = process_kin(Ks,stops,starts,window, norm_length=True,velocity = True,wrt_start=True)
separability_timecourse(Ks,tokens,LD=3)
```

```python
#Feature correlations:
feature_correlations(Ks,feats,tokens,fig_filename =None,window=np.array([-1,2]))
```

```python
#LDA Scatter
Ks = [K[d].copy() for d in tokens]
Ks = process_kin(Ks,stops,starts,window,numTongue=5, norm_length=True, velocity = False, wrt_start=True)
mid = np.arange(37,42)
X = pd.DataFrame(np.nanmean(Kall[0:7,mid,:],axis=1).transpose(),index=notes)
X = X.fillna(X.median(),inplace=True)
#fig_filename = save_pth + subj + '_Kin_LD1LD2.png'
d_proj = LDA_proj_scatter(X,sc = True)
```

```python
#Classification
Ks = [K[d].copy() for d in tokens]
Ks = process_kin(Ks,stops,starts,window,numTongue=50, norm_length=True, velocity = True, wrt_start=True)
#fig_filename = save_pth + subj + '_HG_SVMC.png'
[p,vow_perf] = vowel_svm(Ks,tokens,anat,ncomp = 30,penalty = .001,boot = 100,dt = 'kin',fig_filename=Nonedrange = np.arange(27,32))
```
