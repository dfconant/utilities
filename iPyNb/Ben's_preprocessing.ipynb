```python
#initialize
%load_ext autoreload
%autoreload 2
%matplotlib inline
import numpy as np
import scipy as sp
import pandas as pd
from ECoGstats import run_makeD
from parseTranscripts import parseTextGrid
import os.path
import cPickle as pickle
import ecog_viz as ev
import seaborn as sns
sns.set_style('dark')
align_window = np.array([-1,1])
```

```python
def makeD_multi (pth,blocks,tokens):
    subject = pth.split("/")[-2]
    fname = pth + subject + str(blocks) + str(tokens) + '.p'

    if not os.path.isfile(fname): 

        D = dict((el,np.array([])) for el in tokens)
        mean_D = dict((el,np.array([])) for el in tokens)
        for iblock, block in enumerate(blocks):
            print 'Processing block ' + str(block)
            blockname = subject + '_B' + str(block)
            blockpath = pth + blockname + '/'
            #####
            #convert parseout to dataframe
            textgrid_path = blockpath + blockname + '_transcription_final.TextGrid'
            parseout = parseTextGrid(textgrid_path)
            datamat = [parseout[key] for key in parseout.keys()]
            df = pd.DataFrame(np.vstack(datamat).T,columns=parseout.keys())
            df = df[df['tier'] == 'word']

            # get rid of superfluous columns
            df = df[['label','start', 'stop']]

            # fix common transcription error
            #df['label'][df['label'] == 'EE1'] = 'IY1'
            #df['label'][df['label'] == 'EE2'] = 'IY2'

            # pull mode from label and get rid of number
            df['mode'] = ['speak' if l[-1] == '2' else 'listen' for l in df['label']]
            df['label'] = df['label'].apply(lambda x: x[:-1]) 

            df['label'] = df['label'].astype('category')
            df['mode'] = df['mode'].astype('category')

            df['mid_times'] = (df['start'] + df['stop'])/2

            #df['order'] = mod(range(df.shape[0]),5) + 1
            df['block'] = block
            df['subject'] = subject

            df.to_pickle(blockpath + 'event_df')
            #####

            for this_label in tokens:
                event_times = df['start'][(df['label'] == this_label) & (df['mode'] == 'speak')]
                D[this_label] = np.dstack((D[this_label],run_makeD(blockpath,event_times,
                        align_window))) if D[this_label].size else run_makeD(blockpath,event_times,align_window)
                mean_D[this_label] = np.vstack((mean_D[this_label],np.transpose(np.nanmean(np.log(D[this_label]),
                        axis=1)))) if mean_D[this_label].size else np.transpose(np.nanmean(np.log(D[this_label]),axis=1))

        pickle.dump(D,open(fname,'wb'))

    else:
        print 'File found; Loading...'
        D = pickle.load( open(fname,'rb'))
        print 'Loaded'

    anat = ev.load_anatomy(pth)
    return(D, anat)
```

```python
pth = '/Users/david_conant/Documents/EC84/'
blocks = [65,78,82,85,93,95]
tokens = ['HAWED','HEED','WHOOD']

[D,anat] = makeD_multi(pth,blocks,tokens)
Ds = [D[d] for d in tokens]
ax = ev.plotERPs(Ds,align_window = np.array([-1,1]), anatomy = anat,labels = tokens)
```
