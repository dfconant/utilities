```python
CV = 'split'
boot = 10
ncomp = 86
nTP = 200
alphas = np.logspace(-2,15,20)

window = np.array([-1, 2])
[E,anat,stop_times,start_times] = makeD_multi(pth,blocks,tokens,align_window = window, dtype='HG')
Es = [E[d] for d in tokens]
elects = np.concatenate((np.where(anat=='preCG')[0],np.where(anat=='postCG')[0]))

Fr2 = np.empty((Fs[0].shape[0]+1,boot,nTP,len(alphas)))
Fr2null = np.empty((Fs[0].shape[0]+1,boot,nTP,len(alphas)))

prog = ProgressBar(nTP)
for indt,t in enumerate(np.linspace(200,600,nTP)):
    erange = np.arange(t,t+4)   #Try playing with window size

    #Organize data into X and y
    X,y,yrnd = Ds2Xy(Es,Fs,dt1='HG',dt2='form',ncomp = ncomp,erange = erange.astype(int),elects = elects,frange=np.arange(105,115),krange=np.arange(37,42))

    #Ridge regression
    for inda,a in enumerate(alphas):
        Fr2[:,:,indt,inda],yhat,score = Ridge_MLR(X, y, alpha = a, CV = CV, boot = boot)
        Fr2null[:,:,indt,inda],yhat,score = Ridge_MLR(X, yrnd, alpha = a, CV = CV, boot = boot)
    prog.animate(indt)
R2viz(Fr2,Fr2null,ffeats,b=True)
```
