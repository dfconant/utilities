###Does aligning to movement onset significantly change results?

Because it is likely that the activity we see in vSMC is related to movement, not acoustics, aligning to the onset of the acoustics is causing some degree of noise. In the worst case, if the onset of movement is highly variable wrt the onset of acoustics, the results we're seeing could be really smeared in time (e.g. sharp peaks become slow rising hills). To assess the degree to which this is true, I manually defined the onset of movement for one subject and examined the variability of movement onset, and the effect of aligning to movement onset on ERPs.

One note on defining the onset of movement: it's more complicated than it seems. The main issue I'm having right now is that for several of the mid-vowels, the movement is very subtle (i.e. the tongue doesn't move much from starting position). I'm trying a few ways of best identifying the onset, but for now, I marked some trials (about 1/4) as 'poor_alignment' and set their onset at the acoustic onset. This skews the results a bit, but is an easier comparison to the acoustically aligned data.

Mostly the difference is that everything is moved slightly in time. There are a couple of electrodes that show additional structure (68 and 27) but for the most part, no major changes. I'm pretty surprised by this, given some trials moved almost half a second. Interestingly, there is a lot of activity prior to movement onset (assuming that activity isn't due to the 1/4 of trials I wasn't able to realign), which bodes well for future analyses on movement planning.

Conclusions:
    
    -At least preliminarily, there aren't major differences aside from timing between acoustic and movement onset alignment. If we don't see major differences on ERPs, it's unlikely we'll see differences on the MDS or LDA analyses as they are derived from average activity.
