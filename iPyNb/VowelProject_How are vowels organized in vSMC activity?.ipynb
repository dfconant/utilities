```python
from ECoGstats import run_makeD
from parseTranscripts import parseTextGrid
import os.path
from util import save_table_file
import tables
def makeD_multi (pth,blocks,tokens,align_window = np.array([-1,1]),dtype = 'HG'):
    subject = pth.split("/")[-2]
    fname = pth + subject + str(blocks) + str(tokens) + 'HG.hf5'

    if not os.path.isfile(fname): 

        D = dict((el,np.array([])) for el in tokens)
        mean_D = dict((el,np.array([])) for el in tokens)
        for iblock, block in enumerate(blocks):
            print 'Processing block ' + str(block)
            blockname = subject + '_B' + str(block)
            blockpath = pth + blockname + '/'
            #####
            #convert parseout to dataframe
            textgrid_path = blockpath + blockname + '_transcription_final.TextGrid'
            parseout = parseTextGrid(textgrid_path)
            datamat = [parseout[key] for key in parseout.keys()]
            df = pd.DataFrame(np.vstack(datamat).T,columns=parseout.keys())
            df = df[df['tier'] == 'word']

            # get rid of superfluous columns
            df = df[['label','start', 'stop']]

            # fix common transcription error
            #df['label'][df['label'] == 'EE1'] = 'IY1'
            #df['label'][df['label'] == 'EE2'] = 'IY2'

            # pull mode from label and get rid of number
            df['mode'] = ['speak' if l[-1] == '2' else 'listen' for l in df['label']]
            df['label'] = df['label'].apply(lambda x: x[:-1]) 

            df['label'] = df['label'].astype('category')
            df['mode'] = df['mode'].astype('category')

            df['mid_times'] = (df['start'] + df['stop'])/2

            #df['order'] = mod(range(df.shape[0]),5) + 1
            df['block'] = block
            df['subject'] = subject

            df.to_pickle(blockpath + 'event_df')
            #####

            for this_label in tokens:
                event_times = df['start'][(df['label'] == this_label) & (df['mode'] == 'speak')]
                D[this_label] = np.dstack((D[this_label],run_makeD(blockpath,event_times,
                        align_window))) if D[this_label].size else run_makeD(blockpath,event_times,align_window)
                mean_D[this_label] = np.vstack((mean_D[this_label],np.transpose(np.nanmean(np.log(D[this_label]),
                        axis=1)))) if mean_D[this_label].size else np.transpose(np.nanmean(np.log(D[this_label]),axis=1))

        
        save_table_file(fname,dict(D))

    else:
        print 'File found; Loading...'
        tf = tables.openFile(fname)
        D = {}
        for n in tokens:
            D[n] = tf.root.__getattr__(n).read()
        tf.close()
        print 'Loaded'

    anat = ev.load_anatomy(pth)
    return(D, anat)
```

```python
from BDutils import nans
pth = '/Users/david_conant/Documents/EC41/'
blocks = [19,20,22,27]
#def euclid_MDS(pth,blocks):
#TODO: USE PCA FIRST
#Load Data
tokens = ['AAA','AEE','AHH','EHH','ERR','IHH','IYY','UHH','UWW']
[D,anat] = makeD_multi(pth,blocks,tokens)
Ds = [D[d] for d in tokens]
Dall = np.concatenate(Ds,axis=2)
notes = []
for ind,n in enumerate(tokens):
    notes.append([n]*Ds[ind].shape[2])
notes = np.concatenate(notes,axis=1)
chans = str(np.arange(1,256))
vSMC = np.concatenate((np.where(anat=='preCG')[0],np.where(anat=='postCG')[0]))
Dall = Dall[vSMC,:,:]

means = nans((len(Ds),Ds[0].shape[0],Ds[0].shape[1]))
stes = nans((len(Ds),Ds[0].shape[0],Ds[0].shape[1]))
numTrials = np.empty((len(Ds),1))
for iD,D in enumerate(Ds):
    means[iD,:,:] = np.nanmean(D,axis=2)
    stes[iD,:,:] = np.nanstd(D,axis=2)/np.sqrt(D.shape[2])
    numTrials[iD] = D.shape[2]
pos_e = []


e_dist = np.empty((256,9,9))
for e in range(256):
    e_dist[e,:,:] = spd.squareform(spd.pdist(means[:,e,:]))
    
#e_ranges =[np.arange(175,185),np.arange(195,205),np.arange(115,125)]    
for ind,re in enumerate(e_ranges):

    e_proj = pd.DataFrame(np.nanmean(Dall[:,re,:],axis = 1).transpose(), index = notes)
    e_proj = e_proj.fillna(e_proj.median(),inplace=True)
    e_proj = e_proj.dropna(axis =1,how='any')
    e_centroids = e_proj.groupby(e_proj.index).aggregate(np.median)

    #Distance Matrices
    e_dist = spd.squareform(spd.pdist(e_centroids.values[:,0:2]))

    #MDS
    mds = manifold.MDS(n_components=2, max_iter=10000,dissimilarity = 'precomputed')
    pos_e.append(mds.fit_transform(e_dist))

sns.set_style('white')
sns.set_context('talk')
sns.set_palette('Set1', n_colors = 9)

fig = plt.figure(figsize = (20,10))
gs1 = gridspec.GridSpec(2,4)


ax4 = plt.subplot(gs1[1:2,1:2])
ax4.set_ylabel('vSMC MDS2')
ax4.set_xlabel('vSMC MDS1')
ax4.set_title('-100 ms')
for ind,n in enumerate(np.unique(notes)):
    ppl.scatter(pos_e[0][ind,0],pos_e[0][ind,1], alpha = 1, s=60)
    annotate(n, (pos_e[0][ind,0],pos_e[0][ind,1]))

ax5 = plt.subplot(gs1[1:2,2:3])
ax5.set_xlabel('vSMC MDS1')
ax5.set_title('0 ms')
for ind,n in enumerate(np.unique(notes)):
    ppl.scatter(pos_e[1][ind,0],pos_e[1][ind,1], alpha = 1, s=60)
    annotate(n, (pos_e[1][ind,0],pos_e[1][ind,1]))

ax6 = plt.subplot(gs1[1:2,3:4])
ax6.set_xlabel('vSMC MDS1')
ax6.set_title('+100 ms')
for ind,n in enumerate(np.unique(notes)):
    ppl.scatter(pos_e[2][ind,0],pos_e[2][ind,1], alpha = 1, s=60)
    annotate(n, (pos_e[2][ind,0],pos_e[2][ind,1]))
```

MDS finds it's axes according to a function that is analagous to PCA. It's possible that the dimensions of greatest variability are not the best way to visualize this organization. Another way would be to perform LDA instead, which will find the dimensions of optimal separability between classes. Perhaps this will give us a more consistent organization?

```python
def LDA_proj(pth,blocks,elects):
#Load Data
    tokens = ['AAA','AEE','AHH','EHH','ERR','IHH','IYY','UHH','UWW']
    [D,anat] = makeD_multi(pth,blocks,tokens)
    Ds = [D[d] for d in tokens]
    Dall = np.concatenate(Ds,axis=2)
    notes = []
    for ind,n in enumerate(tokens):
        notes.append([n]*Ds[ind].shape[2])
    notes = np.concatenate(notes,axis=1)
    chans = str(np.arange(1,256))
    
    Dall = Dall[elects,:,:]

    e_ranges =[np.arange(360,370),np.arange(495,505),np.arange(430,440)]
    pos_e = []
    pos_e_med = []
    for ind,re in enumerate(e_ranges):

        e_proj = pd.DataFrame(np.nanmean(Dall[:,re,:],axis = 1).transpose(), index = notes)
        e_proj = e_proj.fillna(e_proj.median(),inplace=True)
        e_proj = e_proj.dropna(axis =1,how='any')
        e_centroids = e_proj.groupby(e_proj.index).aggregate(np.median)

        #Distance Matrices
        e_dist = spd.squareform(spd.pdist(e_centroids.values[:,:2]))

        #LDA
        clf = LDA()
        pos_e.append(pd.DataFrame(clf.fit_transform(e_proj.values, e_proj.index)[:,:2],index = e_proj.index))
        pos_e_med.append(pos_e[-1].groupby(pos_e[-1].index).aggregate(np.median))

    sns.set_style('white')
    sns.set_context('talk')
    sns.set_palette('Set1', n_colors = 9)

    fig = plt.figure(figsize = (20,10))
    gs1 = gridspec.GridSpec(2,4)


    ax4 = plt.subplot(gs1[1:2,1:2])
    ax4.set_ylabel('vSMC LD2')
    ax4.set_xlabel('vSMC LD1')
    ax4.set_title('-100 ms')
    ax = [];
    for n in np.unique(notes):
        ax.append(ppl.scatter(pos_e_med[0].loc[n,0],pos_e_med[0].loc[n,1], alpha = 1, s=60))
        annotate(n, (pos_e_med[0].loc[n,0],pos_e_med[0].loc[n,1]))
    #ppl.legend(ax,np.unique(d.index), bbox_to_anchor=(1, 1))
    
    ax5 = plt.subplot(gs1[1:2,2:3])
    ax5.set_xlabel('vSMC LD1')
    ax5.set_title('0 ms')
    ax = [];
    for n in np.unique(notes):
        ax.append(ppl.scatter(pos_e_med[1].loc[n,0],pos_e_med[1].loc[n,1], alpha = 1, s=60))
        annotate(n, (pos_e_med[1].loc[n,0],pos_e_med[1].loc[n,1]))
    #ppl.legend(ax,np.unique(d.index), bbox_to_anchor=(1, 1))
    
    ax6 = plt.subplot(gs1[1:2,3:4])
    ax6.set_xlabel('vSMC LD1')
    ax6.set_title('+100 ms')
    ax = [];
    for n in np.unique(notes):
        ax.append(ppl.scatter(pos_e_med[2].loc[n,0],pos_e_med[2].loc[n,1], alpha = 1, s=60))
        annotate(n, (pos_e_med[2].loc[n,0],pos_e_med[2].loc[n,1]))
    #ppl.legend(ax,np.unique(d.index), bbox_to_anchor=(1, 1))
```

Conclusions:

    -Organization is not consistent with the acoustics, not consistent across subjects, and evolves over time.
    -There do seem to be some consistencies with the acoustics, but it's hard to convince yourself that it's not just chance.
    -Possible explanations: 1) There is no reason that the neural space should necessarily mirror the acoustic space. The signals associated with moving the articulators could reflect features totally orthogonal to the end point positions/acoustics 2) There is also reason to believe that there could be a lot of variability between subjects. We know that the actual behavior has a lot of inter-subject variability, and that could be even greater at the neural level. Individuals have drastically different vocal tracts, but are trying to hit relatively similar acoustic/kinematic targets. Neural variability across subjects could reflect the fact that the movement constraints vary greatly, even if end points don't as much.
