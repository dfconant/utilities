```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_style('white')
sns.set_palette('Greys_d')
```

# Auto Dataset

---

### Looking at all features

```python
auto = pd.read_csv('auto_mpg.csv',delim_whitespace=True)
origins = np.array(['North America','Europe','Asia'])
auto['origin'] = origins[auto['origin'].values-1]
sns.pairplot(auto)
```

### Zoom in on one relationship: horsepower vs MPG

```python
plt.figure(figsize=(7,6))
plt.scatter(auto['horsepower'],auto['mpg'],c='k')
plt.ylim([0,50])
plt.xlim([0,250])
plt.xlabel('Horsepower')
plt.ylabel('MPG')
```

### Other ways of handling space?

#### Smooth over space

```python
plt.figure(figsize=(7,6))
sns.kdeplot(auto['horsepower'],auto['mpg'],shade=True,cmap='Greys')
plt.ylim([0,50])
plt.xlim([0,250])
plt.xlabel('Horsepower')
plt.ylabel('MPG')
```

#### Bin over space

```python
plt.figure(figsize=(7,6))
plt.hexbin(auto['horsepower'],auto['mpg'], gridsize=20,cmap='Greys')
plt.ylim([0,50])
plt.xlim([0,250])
plt.xlabel('Horsepower')
plt.ylabel('MPG')
```

## How can we improve the visualization?

### Free Parameters: 
* Transparancy 
* Color
* Size

##### Address overlapping values with transparancy

```python
plt.figure(figsize=(7,6))
plt.scatter(auto['horsepower'],auto['mpg'],alpha=0.3,s=60,c='k')
plt.ylim([0,50])
plt.xlim([0,250])
plt.xlabel('Horsepower')
plt.ylabel('MPG')
```

##### Include another continuous variable with size

```python
plt.figure(figsize=(7,6))
sizes = np.power(np.true_divide(auto['weight'],auto['weight'].max()),2)*150
plt.scatter(auto['horsepower'],auto['mpg'],alpha=0.3,s=sizes,c='k')
plt.ylim([0,50])
plt.xlim([0,250])
plt.xlabel('Horsepower')
plt.ylabel('MPG')

sizes = np.linspace(14, 150, 5).astype('int')
wrange = np.linspace(1600,5100,5).astype('int')
labels = ['Weight={0}'.format(s) for s in wrange]
points = [plt.scatter([], [],  s=s, c='gray') for s in sizes]
plt.legend(points, labels, scatterpoints=1)
```

##### Or we could use color instead

```python
plt.figure(figsize=(7,6))
sns.lmplot(x='horsepower',y='mpg',data=auto,hue='weight',fit_reg=False,palette='Greys',legend=False,scatter_kws={'s':60,'alpha':0.6})

colors = sns.color_palette('Greys',5)
wrange = np.linspace(1600,5100,5).astype('int')
labels = ['Weight={0}'.format(s) for s in wrange]
points = [plt.scatter([], [],  s=60, c=c, ) for c in colors]
plt.legend(points, labels, scatterpoints=1)
```

##### Include another categorical variable with color

```python
plt.figure(figsize=(7,6))
sns.lmplot(x='horsepower',y='mpg',data=auto,hue='origin',fit_reg=False,palette='deep',scatter_kws={'s':60,'alpha':0.6})
```
